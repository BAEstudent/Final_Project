In this section I took the "Лента" supermarket chain websided and used Selenium to collect the addresses of all shops of the chain. The link to the website: https://lenta.com/allmarkets/

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

### The chunk of code below collects addresses and links of all 'Lenta' shops; ###

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install())) 
driver.get("https://lenta.com/allmarkets/")
elements = driver.find_elements(by=By.TAG_NAME, value="li")
addresses = list(map(lambda x: x.text, elements))
```

In this section I use the same "News Popularity in Multiple Social Media Platforms" dataset to perform some SQL queries.

```python
import sqlite3
conn = sqlite3.connect("database.sqlite")
c = conn.cursor()
```

```python
news
```

Here I download the dataframe into the sql database

```python
news.to_sql("news", conn, if_exists="append", index=False)
```

```python
c.execute("""
SELECT Title FROM news
where (Source='Bloomberg') AND (SentimentTitle=0) AND (SentimentHeadline=0)
""").fetchall()
```

Here I search for all topics that have "USA TODAY" as a source and have average SentimentHeadline above 0

```python
c.execute("""
SELECT Topic FROM news
WHERE Source='USA TODAY'
GROUP BY Topic
HAVING AVG(SentimentHeadline)>0
""").fetchall()
```
