```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

### The chunk of code below collects addresses and links of all 'Lenta' shops; ###

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install())) 
driver.get("https://lenta.com/allmarkets/")
elements = driver.find_elements(by=By.TAG_NAME, value="li")
addresses = list(map(lambda x: x.text, elements))
links = driver.find_elements(by=By.TAG_NAME, value="a")
req_links = links[-(len(addresses)+29):-34]
links_str = list(map(lambda x: x.get_attribute("href"), req_links))
```

```python
from time import sleep
```

```python
shop_info = {}
for button in req_links:
    button.click()
    sleep(1)
    shop_elements = driver.find_elements(by=By.TAG_NAME, value="li")
    #shop_info_elements = list(map(lambda x: x.text, elements))
    shop_info[button] = shop_elements
    driver.back()
    sleep(1)
```
